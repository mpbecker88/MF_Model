```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from math import erf
from mpmath import mp
from scipy import integrate
from scipy.integrate import simpson
from scipy.special import erf
from dataclasses import dataclass, field
from sklearn.metrics import mean_squared_error

mp.dps = 550

# Customize matplotlib
matplotlib.rcParams.update(
	{
		'text.usetex': False,
		'font.family': 'stixgeneral',
		'mathtext.fontset': 'stix',
		'font.size': 22,
		'ytick.color' : 'black',
		"xtick.color" : 'black',
		'axes.labelcolor' : 'black',
		'axes.edgecolor' : 'black'
	}
)
```

```python
@dataclass
class Parameters:
    Vth:float
    Vr: float
    KE: int
    KI: int
    wE: float
    wI: float
    tauE: float
    tauI: float
    tauL: float
    tauR: float
    EE: float
    EI: float
    EL: float
    nu: float

    muE: float = field(init=False)
    muI: float = field(init=False)
    sigmaE: float = field(init=False)
    sigmaI: float = field(init=False)
    mu: float = field(init=False)
    tau: float = field(init=False)
    hE: float = field(init=False)
    hI: float = field(init=False)

    def __post_init__(self):
        self.recalculate()

    def recalculate(self):
        self.muE = self.wE*self.KE*self.nu*self.tauE
        self.muI = self.wI*self.KI*self.nu*self.tauI
        self.sigmaE = self.wE*np.sqrt(self.KE*self.nu*self.tauE)
        self.sigmaI = self.wI*np.sqrt(self.KI*self.nu*self.tauI)
        self.tau = self.tauL/(1 + self.muE + self.muI)
        self.mu = self.tau*(self.EL + self.muE*self.EE + self.muI*self.EI)/self.tauL
        self.hE = (self.tau/self.tauL)*self.sigmaE*(self.EE - self.mu)
        self.hI = (self.tau/self.tauL)*self.sigmaI*(self.EI - self.mu)

    def set_tauE(self, tauE):
        self.tauE = tauE
        self.recalculate()

    def set_nu(self, nu):
        self.nu = nu
        self.recalculate()

    def calculate_sigma(self):
        return np.sqrt((self.tauE * self.hE**2)/(self.tau + self.tauE) + (self.tauI * self.hI**2)/(self.tau + self.tauI))
```

```python
data_directory_path = "../../Data/V_data/"

df_varV_iRate = pd.read_csv(f"{data_directory_path}var_V_ta_iRate.dat", index_col=0)
df_varV_wE = pd.read_csv(f"{data_directory_path}var_V_ta_wE.dat", index_col=0)
df_varV_wI = pd.read_csv(f"{data_directory_path}var_V_ta_wI.dat", index_col=0)

df_meanV_iRate = pd.read_csv(f"{data_directory_path}mean_V_ta_iRate.dat", index_col=0)
df_meanV_wE = pd.read_csv(f"{data_directory_path}mean_V_ta_wE.dat", index_col=0)
df_meanV_wI = pd.read_csv(f"{data_directory_path}mean_V_ta_wI.dat", index_col=0)

df_varV_wE
```

```python
vec_tauE = np.linspace(0.1, 100, 100)
vec_iRate = np.array([0.005, 0.01, 0.02])
vec_wE = np.array([0.1, 1.0, 10])
vec_wI = np.array([0.1, 1.0, 10])
dic_varV_iRate = {vec_iRate[0]:np.zeros_like(vec_tauE), vec_iRate[1]:np.zeros_like(vec_tauE), vec_iRate[2]:np.zeros_like(vec_tauE)}
dic_varV_wE = {vec_wE[0]:np.zeros_like(vec_tauE), vec_wE[1]:np.zeros_like(vec_tauE), vec_wE[2]:np.zeros_like(vec_tauE)}
dic_varV_wI = {vec_wI[0]:np.zeros_like(vec_tauE), vec_wI[1]:np.zeros_like(vec_tauE), vec_wI[2]:np.zeros_like(vec_tauE)}
dic_meanV_iRate = {vec_iRate[0]:np.zeros_like(vec_tauE), vec_iRate[1]:np.zeros_like(vec_tauE), vec_iRate[2]:np.zeros_like(vec_tauE)}
dic_meanV_wE = {vec_wE[0]:np.zeros_like(vec_tauE), vec_wE[1]:np.zeros_like(vec_tauE), vec_wE[2]:np.zeros_like(vec_tauE)}
dic_meanV_wI = {vec_wI[0]:np.zeros_like(vec_tauE), vec_wI[1]:np.zeros_like(vec_tauE), vec_wI[2]:np.zeros_like(vec_tauE)}
```

```python
for iRate in vec_iRate:
	for i, tauE in enumerate(vec_tauE):
		parameters = Parameters(Vth=-50, Vr=-60, KE=400, KI=100, wE=0.1, wI=0.4, tauE=tauE, tauI=10, tauL=20, tauR=2, EE=0, EI=-80, EL=-60, nu=iRate)

		dic_varV_iRate[iRate][i] = parameters.calculate_sigma()
		dic_meanV_iRate[iRate][i] = parameters.mu

for wE in vec_wE:
	for i, tauE in enumerate(vec_tauE):
		parameters = Parameters(Vth=-50, Vr=-60, KE=400, KI=100, wE=wE, wI=0.8, tauE=tauE, tauI=10, tauL=20, tauR=2, EE=0, EI=-80, EL=-60, nu=0.005)

		dic_varV_wE[wE][i] = parameters.calculate_sigma()
		dic_meanV_wE[wE][i] = parameters.mu

for wI in vec_wI:
	for i, tauE in enumerate(vec_tauE):
		parameters = Parameters(Vth=-50, Vr=-60, KE=400, KI=100, wE=0.5, wI=wI, tauE=tauE, tauI=10, tauL=20, tauR=2, EE=0, EI=-80, EL=-60, nu=0.005)

		dic_varV_wI[wI][i] = parameters.calculate_sigma()
		dic_meanV_wI[wI][i] = parameters.mu
```
